In this notebook, we'll continue where we left off with notebook 3.3 on visualizing time series data. But since this chapter is about machine learning and predictive modeling, we'll see something different. The whole point of analyzing time series data, is to see how a phenomenon behaves over time. Are sales increasing or decreasing? Is my revenue staying static? and so on. However, one other use for analyzing time series data is to forecast what will happen on future events. Are my sales for next year going to increase or decrease based on my current rhythm? Let's see how we can do  this with the shampoo sales over 3 years dataset. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
shampoo = pd.read_csv("shampoo.csv")
shampoo.head()
```

```python
def Clean_Date(date):
    date = "20" + date + "-01"
    return date

shampoo.Month = shampoo.Month.apply(Clean_Date)
shampoo.head()
```

```python
shampoo.Month = pd.to_datetime(shampoo.Month)
shampoo.head()
```

```python
from matplotlib import rcParams 
def MatplotLib_Params():
    #The size of the figure.
    rcParams['figure.figsize'] = (8, 6)
    #Dots per inch or resolution of the figure.
    rcParams['figure.dpi'] = 150
    #Size of the letters and numbers in the plot.
    rcParams['font.size'] = 14
    
#Remember this?
shampoo = shampoo.set_index("Month")
MatplotLib_Params()
plt.plot(shampoo.values)
plt.show()
```

In this dataset, the sales of shampoo are increasing as time increases. Based on these results we can create somewhat of a forecast of how the arrivals for the next years are going to go. To do this we create an ARIMA model.

## ARIMA 

ARIMA is an acronym that stands for:

**AR** = AutoRegressive (assumes that each unit of time value is related to each other)

**I** = Integrated (Assumes the time series is stationary, that's why I mentioned it so much) 

**MA** = Moving Average. (The rolling mean per observation that we also calculated previously while making the time series stationary.)


First let's visualize the autocorrelation betweeen the different values of the time series. Since we need to assume the parameter **AR** is true, we need to confirm that the number of arrivals at any given point is related to another. We can visualize this by plotting the autocorrelation plot.

```python
from pandas.tools.plotting import autocorrelation_plot
autocorrelation_plot(shampoo)
plt.show()
```

The purpose of this plot is to check for **randomness**. If the plot is concentrated very near 0.00, then the data has no correlation between each other, and there's a probability that it's random. The **lag** on the x-axis stands for the units of time, in this case 36 months. Since our data is not focused on 0.00, and we know that in it's origin it isn't random and follows a very visible trend, we can infer that the data of the time series isn't random. This plot also helps us define the parameters of the ARIMA model:

- **P**: The lag order, or number of lag observations. This is the number of units of time we will take on each set of the time series. This can be inferred from the autocorrelation plot. The autocorrelation is positive for about 12 lag values, being significantly positive for the first 5. We can then take 5 as our value for P.

- **D**: The order of differencing. We saw this on the first notebook about time series, which denotes the difference between one values and its previous value. Depending on the order of differenciation, we may calculate the difference between the second previous value, third, etc. For example:

    If $d=0:  y_t  =  Y_t$

    If $d=1:  y_t  =  Y_t - Y_{t-1}$

    If $d=2:  y_t  =  (Y_t - Y_{t-1}) - (Y_{t-1} - Y_{t-2})  =  Y_t - 2Y_{t-1} + Y_{t-2}$ and so on...
    

- **Q**: Order of moving average. As we also saw in the first notebook about time series, this is the amount of values taken for the calculation of the moving average. This value must be lower than P.

Some examples of ARIMA models are:

- ARIMA(1,0,0) = first-order autoregressive model: if the series is stationary and autocorrelated,it can be predicted as a multiple of its own previous value, plus a constant. $_t  =    +  _1Y_{t-1}$

- ARIMA(0,1,0) = random walk: $_t  =    + Y_{t-1}$ where the $$  is the average period-to-period change in Y. This models the behavior of the test time series, but not in the correct time.

- ARIMA(0,1,1) with constant = simple exponential smoothing with growth: $_t   =    + Y_{t-1}  - _1e_{t-1}$ which uses an error to smooth the model into a curve.

For more, take a look here: https://people.duke.edu/~rnau/411arim.htm

For P, let's take 5, for D let's take 1, and for Q let's take 0. These parameters can be optimized by using cross validation, just like on regular machine learning models. To use ARIMA, we can use the statsmodels library. We can also print a summary of the model training, which gives us information about confidence intervals of the predicted values, and the confidence of each individual lags. What you have to look out here for is the constant coefficient and the standard errors. 

```python
from statsmodels.tsa.arima_model import ARIMA

p,d,q = 5,1,0

model = ARIMA(shampoo, order=(p,d,q)) # pass the data and p,d,q parameters.
model_fit = model.fit(disp=0)
print(model_fit.summary())
```

This resulting model also let's us visualize the residuals or errors of the final model. Since the residual fluctuates between large values, there may be variables that cause uncertainty in the prediction that can't be explained by our model.

```python
residuals = pd.DataFrame(model_fit.resid)
residuals.plot()
plt.show()
```

We can also visualize a density plot of the errors, which tells us the overall distribution of the errors of the model.

```python
residuals.plot(kind='kde')
plt.show()
print(residuals.describe())
```

The distribution of the errors can tell us if the model is over/under predicting. Since the mean of the distribution is not 0, it's not completely normal. In thi case, it's slightly negative, so the model is slightly underpredicting. We can optimize the model by grid searching values of P, D and Q, and keeping the combination that gives us the lowest Root Mean Squared Error (RMSE). Let's do just that in this cell.

```python
from sklearn.metrics import mean_squared_error

train = shampoo[:"2003-06-01"].values
test = shampoo["2003-06-01":].values # 6 Months of test.

best_p,best_d,best_q = -1,-1,-1
best_rmse = 1e10

#Iterate through values of P D and Q, this takes a bit.
for p in range(10):
    for d in range(3):
        for q in range(10):
            #In case a combination is invalid, use an exception.
            try:
                historical_data = [year for year in train]
                predictions = []
                for t in range(len(test)):
                    #Create the model.
                    model = ARIMA(historical_data, order=(p,d,q))
                    model_fit = model.fit(disp=0)
                    #Create the forecast.
                    output = model_fit.forecast()
                    yhat = output[0]
                    predictions.append(yhat)
                    #Compare predicted values with real ones.
                    obs = test[t]
                    historical_data.append(obs)
                #Get the error and compare it.
                error = error = mean_squared_error(test, predictions)
                if error < best_rmse:
                    best_p,best_d,best_q = p,d,q
                    best_rmse = error
            except:
                continue
print("Best values for P, D , Q = {},{},{}".format(best_p,best_d,best_q))
```

Finally, we can train our final model with the best values found and plot the comparison between the real values, the predicted values, the upper bound and the lower bound, as well as the whole time series.

```python
historical_data = [year for year in train]
predictions = []
lower_bound = []
upper_bound = []
for t in range(len(test)):
    #Create the model.
    model = ARIMA(historical_data, order=(best_p,best_d,best_q))
    model_fit = model.fit(disp=0)
    #Create the forecast.
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    lower_bound.append(output[2][0][0])
    upper_bound.append(output[2][0][1])
    #Compare predicted values with real ones.
    obs = test[t]
    historical_data.append(obs)

error = mean_squared_error(test, predictions)
print('Test MSE: %.8f' % error)

# Plot the time series.
last_values = list(range(29,36)) # total lags predicted.
plt.plot(train,color = 'black',label = 'train')
plt.plot(last_values,test,label = 'test')
plt.plot(last_values,predictions, color='red',label = 'prediction')
plt.plot(last_values,lower_bound,color = 'orange',label = 'lower')
plt.plot(last_values,upper_bound,color = 'green',label = 'upper')
plt.legend(loc = 'best')
plt.show()
```

The real values had a big spike in sales in the common months, which was unprecedented for our model. Thus, the prediction was off by a bit on the spikes. If it was closer to the upper bound, then it could have done better. Nonetheless, the values predicted by the model follow the trend that the real data was exhibiting. We can also forecast future values of the model. 

```python
final_model = model = ARIMA(shampoo, order=(best_p,best_d,best_q))
final_model_fit = model.fit(disp=0)
future = final_model_fit.forecast(5)

last_values = list(range(36,41)) # total lags predicted.
predictions = future[0]
lower = [pred[0] for pred in future[2]]
upper = [pred[1] for pred in future[2]]

plt.plot(shampoo.values,color = 'black',label = 'train')
plt.plot(last_values,predictions, color='red',label = 'prediction')
plt.plot(last_values,lower,color = 'orange',label = 'lower')
plt.plot(last_values,upper,color = 'green',label = 'upper')
plt.legend(loc = 'best')
plt.show()
```

So in the end, the model predicts a downward spike, then a rise and then a smoother downwards transition, which resonates with the current trend of the data. One other way of working with forecasting on sequential data is a model called Hidden Markov Models or HMM. 

## Hidden Markov Models

So hidden markov models study a sequence that follows a certain order. So what it does is  take the order of a sequence, and assign a "hidden" state to each observation. After that, it looks at the next observation, and assigns it it's own hidden state. And so on and so forth. After it's finished observing, it can predict which is the next state by using it's **transition matrix**. The transition matrix denotes the probability of going from one state to the next. So if you had this sequence

    UP UP DOWN DOWN UP
    
The transition matrix would be:

$P(UP | UP) = 0.5 $ Since 1 time after it was UP the next value was also UP, while the other time it was UP the next value was down. So 1/(1 + 1) = 1/2 = 0.5

$P( DOWN | UP) = 0.5 same as above.

$P( UP | DOWN) = 0.5 since only once after it's DOWN the next value is UP.

$P(DOWN | DOWN) = 0.5 same as above.

    So $T =
            [0.5 0.5  
             0.5 0.5] 
             

To use HMMs on Python, we first need to install the hmmlearn library. Use pip-install on the conda terminal to install it.

```python
from hmmlearn import hmm
```

After that, let's modify the shampoo dataset so we get the direction the sales are going. So if the sale of this month is greater than the previous then we change the value to UP or 1, else, we change the value to DOWN or 0.

```python
shampoo_directions = shampoo.copy()
directions = []
for i in range(1,len(shampoo_directions)):
    if shampoo_directions.values[i,1] > shampoo_directions.values[i-1,1]:
        directions.append(1)
    else:
        directions.append(0)

shampoo_transitions = pd.DataFrame({"Dates":shampoo_directions.Month[1:],"Direction":directions})
shampoo_transitions
```

Next we create a Multinomial HMM. This one is used for strictly categorical variables, we you don't have a continous distribution.

```python
lr = hmm.MultinomialHMM(n_components=2) # The number of hidden states.
X = shampoo_transitions.Direction.values.reshape(-1,1) #1D array reshape.
lr.fit(X)
```

We can visualize the starting probability for each state of the model after it has been trained.

```python
lr.startprob_
```

So the next value is more likely to be 0 or down, than 1. Let's see the transition matrix. 

```python
lr.transmat_
```

And finally, let's try to predict what the next value of the sequence would be. To do this we use the **emission matrix**. This emission matrix is similar to the transition matrix, but it denotes which **value** is most likely to be next, given a certain hidden state. In our case, we first predict what the next hidden state would be. 

```python
prediction = lr.predict(X)
prediction[-1] #Final prediction. 
```

And after we have the hidden state, we map it to the highest probability of the emission matrix of that given state.

```python
final_value = np.argmax(lr.transmat_[prediction[-1]]) # Get the index of the highest number for the prediction in the t matrix.
np.argmax(lr.emissionprob_[final_value]) # Get the index of the highest emission for that transition.
```

And there we go. Apparently, the next month the sales will go down. If you see the prediction for the time series earlier, this prediction matches.
