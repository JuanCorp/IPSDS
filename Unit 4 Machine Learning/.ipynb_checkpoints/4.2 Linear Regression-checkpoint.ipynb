Let's imagine a situation in which you have lots of data. Maybe something like the population of a city and the profit of the product on that city. Maybe you are a new salesman  with a new product. So the first thing you get is the data for the city.

The process of fitting a line through data points like this is the process of **Linear Regression**. So for a dataset like the one we have now, there exists an equation of the form:

$ {y} = \theta_{0} x_{0} + \theta_{1} x_{1}$

Where $ x_{0} = 1 $

Or for our specific case:

$ {Profit_t} = \theta_{0} + \theta_{1} Population_{t}$

Which means that a Profit can be estimated for any given Population given a set of parameters $\theta_{0}$ and $\theta_{1}$. To maintain consistency with this equation, let's add a new column, equivalent to $x_{0}$ which is filled with 1s.

```python
#Creates a new array filled with 1s.
x0 = np.ones_like(pops.Profit)
pops["x0"] = x0

#rearrange the dataframe
pops = pops[["x0","Population","Profit"]]
pops.head()
```

Now we first define our **hypothesis** function.  This hypothesis is what we initially think the values for the Profits would be, which would be the dot product for the $\theta$ paramater vector and the $x$ matrix. Since on its first column the x matrix is 1, the final calculation would look like this:

$h(x) = \theta_{0} + \theta_{1} Population_{t}$ 

As an example, for an Population of 7 and a theta vector of [0.5, 0.3], the hypothesis function would be equal to:

$h(7) = 0.5 + 0.3(7)  = 21.5$

The one thing to note is that since we are working with *vectors*, this operation would be calculated for each observation in the data.

```python
#Hypothesis function. Multiplies the theta vector with each sample of the X matrix and sums the columns of the result.
def hypothesis_function(X, theta):
    return np.dot(X,theta)
```

This operation returns a vector, where each value is the results of the calculation of the hypothesis function that was detailed above. Let's test it for some arbitrary values of theta. 

```python
# X matrix.
X = pops[['x0','Population']]
X = X.values
#Create theta vector of ones, with the shape of the number of columns in the X matrix.
theta = np.ones(X.shape[1])
hypothesis = hypothesis_function(X, theta)
hypothesis
```

So now we got a rather basic hypothesis of what the Profit values should be. Now, let's calculate how far off the predictions were. This function that calculates how far off each prediction was is the **Cost Function**. It basically measures the performance of the model. What this calculates is the sum of the squared difference of the real value of the Profit and the hypothesized value of the Profit. After this, the result is divided  by twice the size of the dataset. Thus, we would have a vectorized formula like this:

$Cost = \dfrac{1}{2m} (hypothesis(X,theta) - y) * (hypothesis(X,theta) - y)^T $

Which can be simplified to

$Cost = \dfrac{1}{2m} (e)(e')$

Where $e$ is the residual vector result of the difference between the hypothesis and the real values, $e'$ is the transpose of the residuals and $m$ is the size of the dataset. Since we have two vectors, the original and the transpose, the results of their product is divided by twice their size. This cost, or loss function denoted the average difference between the real values and the predicted ones.

```python
def Cost_Function(real,hypothesis):
    #Subtract the two vectors.
    residual = real - hypothesis
    #The amount of data
    size = len(real)
    # The dot product between the two vectors
    product = np.dot(residual.T,residual)
    #Divide their result by two times their size.
    return product/(2*size)


print(Cost_Function(pops.Profit,hypothesis))
```

Let's also plot the predictions against the real values. 

```python
plt.scatter(pops.Population,pops.Profit,color = 'blue')
plt.scatter(pops.Population,hypothesis,color = 'red')
plt.show()
```

The values for the hypothesis are way off from the real values, so the high value for the cost function denoted that our hypothesis was really bad. Let's try calculating the hypothesis with a different value for $\theta$

```python
def Compare_Theta(theta):
    hypothesis = hypothesis_function(X,theta)
    print(Cost_Function(pops.Profit,hypothesis))
    plt.scatter(pops.Population,pops.Profit,color = 'blue')
    plt.scatter(pops.Population,hypothesis,color = 'red')
    plt.show()

Compare_Theta([0.1,0.8])
```

This combination of values for theta is better than the previous one. Try using other combinations of theta on the Compare_Theta function to see which one works best.

## Gradient Descent

Changing the values of theta manually is a long and tedious process, and if we had  a large amount of parameters, then it would be almost impossible to optimize. Since the objective of the model is to minimize the cost function, thus getting the best possible fit for parameters, there must exist a combination of values for $\theta$ which minimizes it. Let's try plotting this behavior by iterating through many values of theta, and plotting how the cost function evolves through as $\theta$ changes.

```python
def Plot_Cost_Change():
    theta = [0.0,0.0]
    changes = []
    for i in np.arange(0.0, 1.0, 0.1):
        for j in np.arange(0.0,1.0,0.1):
            theta = [i,j]
            hyp = hypothesis_function(X,theta)
            changes.append(Cost_Function(pops.Profit, hyp))
    
    plt.plot(range(0,100),changes)
    plt.show()

Plot_Cost_Change()
```

If we obsersve how the function we created rotates its values, we can see that the cost is near its minimum when the value of $\theta_1$ is near 0.5. See how it resets every time the value of $\theta_0$ is reset. Let's try creating a  similar plot, but leaving the value of $\theta_0$ unchanged.

```python
def Plot_Cost_Change_Static():
    theta = [0.0,0.0]
    changes = []
    for i in np.arange(0.0, 2.0, 0.1):
            theta = [0.0,i]
            hyp = hypothesis_function(X,theta)
            changes.append(Cost_Function(pops.Profit, hyp))
    
    plt.plot(np.arange(0.0,2.0,0.1),changes)
    plt.show()

Plot_Cost_Change_Static()
```

So now we can see how the cost is almost minimum near 0.8.  But would we go about finding this minimum value for the cost automatically?  This is where **Gradient Descent** comes along. In Calculus, the gradient points towards the direction of the greatest rate of increase of a variable. In Gradient Descent, we update the parameter vector $\theta$ for several training epochs. After each epoch, we update the  paramaters based on how  well the parameters did on the cost function. Since we want to minimize the value of the cost function, we update the parameters towards the direction *negative* to the gradient. We repeat this procedure until it converges, meaning the  values for the parameters aren't changing anymore. The formula for this is:

$repeat: \theta := \theta - \dfrac{\alpha}{m} (X^T  \cdot (h(x) - y))$
$ until \ convergence$


Where $h(x)$ is the hypothesis function, $\dfrac{1}{m} (X^T  \cdot (h(x) - y))$ is the gradient of the Cost function,  $m$ is the number of examples, and $\alpha$ is the learning rate. The learning rate denotes how aggresive the changes of values of theta are. The higher the learning rate, the greater the changes in the parameter vector $\theta$. However, if the learning rate is too large, then Gradient Descent my not converge. On the other hand, if it's too small, it may take a long time to converge. Let's create a function that performs this process.

```python
def Gradient_Descent(epochs,lrate = 0.01, theta = [0.0,0.0]):
    size = len(pops.Profit)
    cost_history = []
    for i in range(epochs):
        hyp = hypothesis_function(X,theta)
        cost = Cost_Function(pops.Profit, hyp)
        cost_history.append(cost)
        print("Training for epoch " + str(i+1)  + ", theta vector is " + str(theta) + ", cost is:  ", str(cost))
        gradient = (1/size) * X.T.dot(hyp - pops.Profit)
        theta = theta - lrate * gradient
    
    plt.plot(range(0,epochs),cost_history)
    plt.title("Cost function over time")
    plt.show()
    return theta
        
print("Best parameters for theta were " + str(Gradient_Descent(25)))
```

If we set the learning rate to a lower number, we see that it learns slower.

```python
print("Best parameters for theta were " + str(Gradient_Descent(25,lrate = 0.001)))
```

However, if we set it to high, it doesn't improve, and it never converges.

```python
print("Best parameters for theta were " + str(Gradient_Descent(25,lrate = 0.1)))
```

So be sure to experiment with the learning rate until you find one that converges into the minimum. We'll later see how to optimize the learning rate for the data as well. Now that we have a way to optimize the parameter vector $\theta\, let's create a function that predicts the cost of a house with some new Populations.  

```python
best_theta = Gradient_Descent(50)
```

Let's plot how these parameters compare to the real values.

```python
Compare_Theta(best_theta)
```

### Scikit-learn 

Since in a working environment, you'll most likely will not be implementing these algorithms from scratch, I'll show you how to get to the same conclusion with the Python Machine Learning library Scikit-learn. This is a very popular library with very useful functions, which we will regularly see throughout these next units.

```python
# Scikit learn Linear Regression

from sklearn.linear_model import LinearRegression

#Initialize the linear regression model.
lr = LinearRegression()
#Train the model with the populations and profits. No need to add the x0 vector, it already does this for us.
lr.fit(pops.Population.values.reshape(-1,1),pops.Profit.values.reshape(-1,1))

New_Populations = [11,7.25, 6.49, 13.6]
new_pops = np.asarray(New_Populations)
#Predict the profits for the new populations.
Profits = lr.predict(new_pops.reshape(-1,1))

for i in range(0,len(Profits)):
    print("Profit for Population " + str(New_Populations[i]) + " is " + str(Profits[i]))



plt.scatter(pops.Population, pops.Profit, color = 'blue')
plt.scatter(New_Populations, Profits, color = 'red')
plt.show()

sk_predict = lr.predict(pops.Population.values.reshape(-1,1))  
print("Cost for sklearn predictions: " + str(Cost_Function(pops.Profit,sk_predict.flatten())))
print("Theta for sklearn predictions: "+ str([lr.intercept_,lr.coef_]))
```

As we can see, the sklearn model outperforms our handcrafted model. We can match it however by optimizing the learning rate, and/or by increasing the number of epochs trained.
